```python
import numpy as np
import h5py
import matplotlib.pyplot as plt
#%matplotlib widget


B = h5py.File("test_faraday.h5")

By  = B["dBy"][1:-1]
Bz  = B["dBz"][1:-1]

x = np.arange(0.1, 10.1, 0.1)# dual grid

# Analytical solution

k = 2.0
dt = 0.01
By_ana = -k*dt*np.sin(k*x)
Bz_ana = -k*dt*np.cos(k*x)

plt.figure()
plt.plot(x, By, 'bo', label='By PIC code')
plt.plot(x, Bz, 'ro', label='Bz PIC code')
plt.plot(x, By_ana, 'b', label='By theoretical')
plt.plot(x, Bz_ana, 'r', label='Bz theoretical')
plt.xlabel(r'x')
plt.ylabel(r'B-field')
plt.title(r'B$_y$ and B$_z$ for E$_y$(x) = sin(2x), E$_z$(x) = cos(2x)')
plt.legend()
plt.show()
```

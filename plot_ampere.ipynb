```python
import numpy as np
import h5py
import matplotlib.pyplot as plt
#%matplotlib widget


J = h5py.File("test_ampere.h5")

Jy  = J["Jy"][1:-1]
Jz  = J["Jz"][1:-1]

x = np.arange(0.05, 10.15, 0.1) # primal grid

# Analytical solution

k = 2.0
Jy_ana = k*np.sin(k*x)
Jz_ana = k*np.cos(k*x)

plt.figure()
plt.plot(x, Jy, 'bo', label='Jy PIC code')
plt.plot(x, Jz, 'ro', label='Jz PIC code')
plt.plot(x, Jy_ana, 'b', label='Jy theoretical')
plt.plot(x, Jz_ana, 'r', label='Jz theoretical')
plt.xlabel(r'x')
plt.ylabel(r'Current')
plt.title(r'J$_y$ and J$_z$ for B$_y$(x) = sin(2x), B$_z$(x) = cos(2x)')
plt.legend()
plt.show()
```
